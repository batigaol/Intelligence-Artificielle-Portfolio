```python
import numpy as np
import pandas as pd
df=pd.read_csv('Base.csv', sep=" ",header=None)
df.columns = [ "Linear", "Polynom2" ,"Polynom3", "Polynom4","Polynom5", "Multilinear" ,"Log" ,"Exp","Sinus"]
#df.head(10)
df.shape
df.head(10)
```

```python
import numpy as np
import pandas as pd
fakes=pd.read_csv('Fakes_data.csv', sep=" ",header=None)
fakes.columns = ["Unif","Cauchy", "Norm", "X1" ,"X2", "X3","X4", "X5","X6"]
fakes.head(10) 
```

Fonction permettant de creer un sample size systematic 

```python
# Define systematic sampling function
def systematic_sampling(df, step):
 
    indexes = np.arange(0, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample
```

Création d'un sample size systematic de 1 sur 20 sur le dataset base

```python
# Obtain a systematic sample and save it in a new variable
systematic_sample_1_20 = systematic_sampling(df, 20)
systematic_sample_1_20.head(10)
# systematic_sample_1_20.shape
```

Création d'un sample size systematic de 1 sur 100 sur le dataset fakes

```python
# Obtain a systematic sample and save it in a new variable
systematic_sample_Fakes = systematic_sampling(fakes, 20)
#df.head(10)
systematic_sample_Fakes.head(10)
```

```python
## add 10% of gaussian noise to my data 
# x is my training data
# mu is the mean
# std is the standard deviation
import numpy as np
def gaussian_noise(x,mu,std):
    mu=0.0
    std = 0.1 * np.std(x) # for 10% Gaussian noise   
    noise = np.random.normal(mu, std, size = x.shape)
    x_noisy = x + noise
    return x_noisy 
```
