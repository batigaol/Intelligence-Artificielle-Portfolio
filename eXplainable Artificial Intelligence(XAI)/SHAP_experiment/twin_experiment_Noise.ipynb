```python
## add +-10% of  noise to my data 
# x is my training data
# mu is the mean
# std is the standard deviation
import numpy as np
def Noise(x):
    
    percent =[0.1,-0.1,0.3,-0.3,0.5,-0.5,1,-1]
    arr = percent.reshape((1000, 9))
    noise = np.random.permutation(arr)
    x_noisy = x + noise
    return x_noisy 
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv('Base.csv', sep=" ",header=None)
df.columns = [ "Linear", "Polynom2" ,"Polynom3", "Polynom4","Polynom5", "Multilinear" ,"Log" ,"Exp","Sinus"]
df.shape
df.head(10) 
```

```python
df_noisy=Noise(df)
df_noisy.head(10)
```
