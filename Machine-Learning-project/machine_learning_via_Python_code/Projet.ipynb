```python
import numpy as np
import pandas as pd
df=pd.read_csv('Base_donne.txt', sep=" ",header=None)
df.columns = ["S_3.05m", "S_9.45m", "S_1137m" ,"T_3.05m", "T_9.45m", "ADT" ,"Longitude" ,"Latitude","mois"]
df.head(10)
df.shape
  
```

```python
print(df.shape)
print(df.columns)
print(df.dtypes)
print(df.info())
   
```

```python
df.head(10)
```

```python
df = df.iloc[1:]
df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.impute import KNNImputer
#### visualisation des NA
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.savefig('VisualNA.jpg',dpi=800, quality=90, optimize=True, progressive=True) 
#plt.savefig("visualNA.png",dpi=300, bbox_inches='tight')
len(df), len(df.dropna())
```

```python
############## Echantillon de la cross-validation de 5000 donneés
df_cross_validation=df.sample(5000)
print(df_cross_validation.shape)
df_cross_validation.head(10)
```

```python
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
imp = IterativeImputer(estimator=lr,missing_values=np.nan, max_iter=100, verbose=2, 
                       imputation_order='roman',random_state=0)
Df_cross_validation=pd.DataFrame(imp.fit_transform(df_cross_validation))

print(Df_cross_validation.shape)

```

```python
Df_cross_validation.columns = ["S_3.05m", "S_9.45m", "S_1137m" ,"T_3.05m", "T_9.45m", "ADT" ,"Longitude" ,"Latitude","mois"]
Df_cross_validation.head(10)
```

```python
#Visualiser l'ensemble des scatter-plot possibles d'un jeu de données avec la fonction scatter_matrix 
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
scatter_matrix(Df_cross_validation.iloc[:,0:9], alpha = 1, figsize = (9, 9), diagonal = 'kde',color="red")
plt.savefig('VisualVar.jpg',dpi=800, quality=90, optimize=True, progressive=True) 
#plt.savefig("visualvar.png")
#plt.show() # On voit que les variables discrètes sont éliminées
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(Df_cross_validation, hue='S_1137m')
plt.savefig('Visualplot.jpg',dpi=800, quality=90, optimize=True, progressive=True) 
#plt.savefig("visualplot.png")
```

```python
## correlation entre les variables
import seaborn as sns
sns.heatmap(Df_cross_validation.corr())
plt.savefig('Visualcor.jpg',dpi=800, quality=90, optimize=True, progressive=True) 
plt.savefig("visualcor.png")
```

```python
sns.heatmap(Df_cross_validation.corr(), annot=True, cmap='Greens')
plt.title("Matrice des corrélations entre variables\n", fontsize=14, color='#009432')
plt.savefig('VisualCor.jpg',dpi=800, quality=90, optimize=True, progressive=True) 
#plt.savefig("visualCor.png")
```

```python
########### Modele de regression
### preparation du dataset de la cross-validation
#from papierstat.datasets import load_wines_dataset

#X = Df_cross_validation.drop(['T_3.05m', 'T_9.45m ','mois'], axis=1)
y = Df_cross_validation['S_1137m']
################## 
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=10)
###################### 

from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer
from sklearn.metrics import mean_squared_error
#### score de test
score=cross_val_score(knn, Df_cross_validation , y, cv=10, scoring=make_scorer(mean_squared_error))
mean_mse, std_mse = np.mean(score), np.std(score)
print(score)
print("mean MSE: %.2E, std MSE: %.2E" %(mean_mse, std_mse))
# fit the regressor with x and y data
knn.fit(Df_cross_validation, y) 
print("Writing predictions to file...")
df['Predicted'] = pd.Series(knn.predict(Df_cross_validation), index=Df_cross_validation.index)
df.to_csv('pred_fileknn.csv')
```

```python
# Fitting Random Forest Regression to the dataset
# import the regressor
from sklearn.ensemble import RandomForestRegressor
K=10
 # create regressor object
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
Score=cross_val_score(regressor, Df_cross_validation , y, cv=10, scoring=make_scorer(mean_squared_error))
mean_mse, std_mse = np.mean(Score), np.std(Score)
print(Score)
print("mean MSE: %.2E, std MSE: %.2E" %(mean_mse, std_mse))
# fit the regressor with x and y data
regressor.fit(Df_cross_validation, y) 
print("Writing predictions to file...")
df['Predicted'] = pd.Series(regressor.predict(Df_cross_validation), index=Df_cross_validation.index)
df.to_csv('pred_fileRandomForest.csv')
```

```python
############### visualisation de la prediction
import matplotlib.pyplot as plt
for i, modele  in enumerate(["KNeighborsRegressor","RandomForestRegressor"]):
    plt.subplot(2, 1,i+1)
    plt.scatter(y,knn.predict(Df_cross_validation), color="darkorange", label="data")
    plt.plot(y,regressor.predict(Df_cross_validation), color="navy", label="prediction")
    plt.axis("tight")
    plt.legend()
    plt.title("Modéle de Regression(k-folds = %i, modéle = '%s')" % (10, modele))
    #plt.savefig("visualmodele.png")
    plt.savefig('line_plot.jpg',dpi=300, quality=80, optimize=True, progressive=True)  
    plt.tight_layout(pad=2)
    #plt.show()
   
```
